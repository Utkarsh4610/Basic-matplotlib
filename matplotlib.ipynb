```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
x = np.linspace(0,10,20)
x
```

```python
y = np.random.randint(0,50,20)
y
```

```python
# Things which we can apply on plt
dir(plt)
```

```python
plt.plot(y)
```

```python
plt.plot(x,y)
```

```python
plt.plot(x,y,'r')
```

```python
plt.plot(x,y,'r')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Plot')
plt.show()
```

```python
plt.plot(x,y,'c*-')
plt.show()
```

```python
plt.plot(x,y,'c*--')
plt.show()
```

```python
plt.plot(x,y,'co',markersize=2)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(x,y,'r*--')
plt.subplot(2,2,3)
plt.plot(x,y,'mo-')
plt.title('Sub Plot3')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
plt.title('Sub Plot4')
plt.show()
```

```python
plt.figure(figsize=(15,10))
for i in range(1,5):
    plt.subplot(2,2,i)
    plt.plot(x,y)
plt.show()
```

```python
# To get the full control over the ploting style we can use the object oriented way of plotting like matplot
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,'co-')
```

```python
fig = plt.figure()
axes = fig.add_axes([0,0,1,1])
axes.plot(x,y,'r--')
axes.set_xlabel('X Axis')
axes.set_ylabel('Y Axis')
axes.set_title('This is title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0,0,1,1])
axes2 = fig.add_axes([0.05,.55,.4,.4])
axes1.plot(x,np.sort(y),'r')
axes2.plot(x,y,'c')
#fig.tight_layout()
fig.savefig('obj.png',dpi=200,bbox_inches="tight")
```

```python
plt.plot(x,y,'co--',markersize=2,label='Not sorted')
plt.plot(x,np.sort(y),'r',label='Sorted')
plt.legend(loc='upper center')
plt.show()
```

```python
fig,ax = plt.subplots(1,3,figsize=(20,5))
ax[0].plot(x,y)
```

```python
plt.scatter(x,y)
```

```python
plt.bar(x,y,width=.5)
```

```python
fig,ax = plt.subplots(1,2,figsize=(10,4))
ax[0].plot(x,y)
ax[1].plot(x,np.exp(x))
```
